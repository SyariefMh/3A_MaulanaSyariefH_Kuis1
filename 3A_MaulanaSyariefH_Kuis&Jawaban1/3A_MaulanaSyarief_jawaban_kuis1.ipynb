```python
lebel_sex = {
    'male': 1,
    'female': 0
}

lebel_smoker ={
    'yes' : 1,
    'no' : 0
}

lebel_region ={
    'northwest' : 0,
    'southwest' : 1,
    'southeast' : 2,
    'northeast' : 3 
}

# Encode Label (Labeling)
df['sex'] = df['sex'].map(lebel_sex)
df['smoker'] = df['smoker'].map(lebel_smoker)
df['region'] = df['region'].map(lebel_region)


# Check Data Head
df.head()
```

setelah nya kita buat sebuah label kode untuk mempermudah dan mempersingkat.

lalu kita cetak dan hasil nya seperti diatas.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

#Dataset
X = df[['age', 'bmi', 'region','sex', 'smoker', 'children']]
  
# VIF Dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# Calculating
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

selanjutnya kita import variance inflation factor dan hasil nya :
age dan bmi merupakan variabel independen memiliki korelasi yang kuat dengan variabel independen lainnya
sedangkan untuk region, sex, smoker, dan children merupakan variabel independen sedikit memiliki korelasi dengan variabel independen yang lain
